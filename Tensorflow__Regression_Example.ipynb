###Import libraries




```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
print(tf.__version__)
```

```python
import logging
logging.getLogger('tensorflow').disabled = True
```

### Load the dataset

```python
filename = "https://raw.githubusercontent.com/codeforgirls-sa/ml/master/math_cleaned_data.csv"
```

###Read the dataset

```python
df = pd.read_csv(filename)
```

##Data Exploration

```python
df.head()
```

```python
df.info()
```



##Start Model Development



```python
#Define labels and features
labels = df['G3'].values
features = df.drop(['G3'],axis=1).values
```

1. **Spilt data**

```python
#Spilt the data; 30% for testing and 70% for training
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.30)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```


   2. **Bulid, Train and Evaluate the Model**


```python
# Build The structure of Model
def build_model():
    model = keras.Sequential([
      layers.Dense(64, activation='relu', input_shape=X_train.shape),
      layers.Dense(64, activation='relu'),
      layers.Dense(1)
    ])

    optimizer = tf.keras.optimizers.RMSprop(0.001)

    model.compile(loss='mse',
                  optimizer=optimizer,
                  metrics=['mae', 'mse'])
    return model

model = build_model()
```

```python
#Train the model
EPOCHS = 1000
model.fit(X_train, y_train, epochs=EPOCHS, verbose=2)
```

```python
#Evaluate the model
loss, mae, mse = model.evaluate(X_test,  y_test)
print("The mean absolute error: {:5.2f}".format(mae))
print("The mean squared error: {:5.2f}".format(mse))
print("The Loss: {:5.2f}".format(loss))
```

3. **Predict**

```python
# Predict 
predict = model.predict(X_test)
idx= np.random.choice(X_test.shape[0]) 
print ("Predicted final grade : ", predict[idx])
print ("Real final grade : ", y_test[idx])
```

4. **Visualize**

```python
# visualization
plt.figure(figsize=(10,10))
plt.scatter(y_test, predict,  color='Grey')
plt.scatter(y_test[idx], predict[idx] , color='blue')
plt.xlabel('Real final grade')
plt.ylabel('Predicted final grade')
plt.plot([0, 30], [0, 30], color='red')
```

#Save, Load and Use the Model

```python
model.save("saved_model")
```

```python
saved_model= tf.keras.models.load_model('saved_model') 
```

```python
#check if it the same model by print the score for test set
n_loss, n_mae, n_mse = saved_model.evaluate(X_test,  y_test)
print("The mean absolute error: {:5.2f}".format(n_mae))
print("The mean squared error: {:5.2f}".format(n_mse))
print("The Loss: {:5.2f}".format(n_loss))
```
